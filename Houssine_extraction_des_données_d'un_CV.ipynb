###1-Lecture du fichier de type PDF


```
import PyPDF2 as pdf
```

```
file = open(r'sample_data/houssine.pdf', 'rb')
```

```
file
```

```
pdf_reader = pdf.PdfFileReader(file)
```

```
pdf_reader
```

```
# check if the pdf file is Encrypted or no 
pdf_reader.getIsEncrypted()
```

```
page1=pdf_reader.getPage(0)
```

```
text = page1.extractText()
```

```
print(text)
```

###2-Prétraitement du text


```
import re
#elimination des espaces
text = re.sub(' +', ' ', text)
#elimination des lignes vides
text = re.sub(r'\n\s*\n','\n',text,re.MULTILINE)
```

```
print(text)
```

###3-Tokenizing

```
import spacy
nlp = spacy.load('en')
```

```
doc = nlp(text)
```

```
print(f"Token \t\tLemma \t\tStopword".format('Token', 'Lemma', 'Stopword'))
print("-"*40)
for token in doc:
    print(f"{str(token).strip()}\t\t{token.lemma_.strip()}\t\t{token.is_stop}")
```
